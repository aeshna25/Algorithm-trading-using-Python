# Cummalative  returns and Dropdowns etc..

```python
apple = yf.download('AAPL')
```

```python
apple = apple.Close.to_frame()
```

```python
apple['d_returns'] = np.log(apple.div(apple.shift(1)))
```

```python
apple = apple.dropna()
```

```python
# sum of the returns
apple['d_returns'].sum()
```

```python
# what if we invested $1 on the first day, what will be the amount today
np.exp(apple['d_returns'].sum())
```

```python
# cummalative returns
apple['cummareturns'] = apple['d_returns'].cumsum().apply(np.exp)  # it should give the same result as above on the last cell
```

```python
apple
```

```python
apple['cummareturns'].plot(figsize=(10,6), title ='Buy and hold : APPLE')
plt.legend(loc='upper left',fontsize=15)
plt.show()
```

1 Dollor invested in 1980, should return you around 1700 Dollor in 2024

```python
annual_avg_return= apple['d_returns'].mean()*252
annual_avg_return
```

```python
annual_std= apple['d_returns'].std()*np.sqrt(252)
annual_std
```

```python
# how to calculate drawdowns
apple['cummax'] = apple['cummareturns'].cummax()
apple
```

```python
apple[['cummareturns','cummax']].plot(figsize=(10,6), title ='Buy and hold + Cummax: APPLE')
plt.legend(loc='upper left',fontsize=15)
plt.show()
```

```python
apple['drowdown'] =apple['cummax'] - apple['cummareturns']
apple
```

```python
apple['drowdown'].max()
```

```python
apple['drowdown'].idxmax()
```

```python
apple.loc[apple['drowdown'].idxmax()]
```

```python
apple.loc[apple.index <= '2023-01-05']
```

```python
apple['drawdown%'] = (apple['cummax'] - apple['cummareturns']) / apple['cummax']
apple
```

```python
apple['drawdown%'].max()
```

```python
apple['drawdown%'].idxmax()
```

## SMA Strategy

```python
data = apple.loc[apple.index >= '1991-01-01'][['Close']]
data
```

```python
SMA_s=50
SMA_l =200

data['SMA_s'] = data['Close'].rolling(window=SMA_s, min_periods=SMA_s).mean()
data['SMA_l'] = data['Close'].rolling(window=SMA_l, min_periods=SMA_l).mean()
data
```

```python
data.plot(figsize=(10,6), title = "AAPL - SMA{} | SMA{}".format(SMA_s, SMA_l))
plt.legend(loc='upper left',fontsize=15)
plt.show()
```

```python
data.loc['2020'].plot(figsize=(10,6), title = "AAPL - SMA{} | SMA{}".format(SMA_s, SMA_l))
plt.legend(loc='upper left',fontsize=15)
plt.show()
```

```python
data.loc['2019'].plot(figsize=(10,6), title = "AAPL - SMA{} | SMA{}".format(SMA_s, SMA_l))
plt.legend(loc='upper left',fontsize=15)
plt.show()
```

```python
data = data.dropna()
data
```

```python
# sma_s > sma_l : long(1) else short(-1)

data['Position'] = np.where(data['SMA_s'] > data['SMA_l'], 1, -1)
data
```

```python
data.loc['2019',["SMA_s","SMA_l","Position"]].plot(figsize=(10,6), title = "AAPL - SMA{} | SMA{}".format(SMA_s, SMA_l), secondary_y = "Position")
plt.legend(loc='upper left',fontsize=15)
plt.show()
```

```python
data['returnsbh'] = np.log(data['Close'].div(data['Close'].shift(1))) # buy and hold
data['strategy'] = data['Position'].shift(1) * data['returnsbh']
data
```

```python
data= data.dropna()
data.head()
```

```python
data[['returnsbh','strategy']].sum()
```

Buy and hold performs better

```python
data[['returnsbh','strategy']].sum().apply(np.exp)
```

dollor 1 investment will yeild dollar 447 using buy and hold and yeilds dollor 7 with strategy

```python
data[['returnsbh','strategy']].std()*np.sqrt(252) #annual std
```

Both have similar variance

## Strategy adjusted with long bias

```python
data['Position2'] = np.where(data['SMA_s'] > data['SMA_l'], 1, 0)
data
```

```python
data['strategy2'] = data['Position2'].shift(1) * data['returnsbh']
data
```

```python
data = data.dropna()
data.head()
```

```python
data[['returnsbh','strategy','strategy2']].sum()
```

```python
data[['returnsbh','strategy','strategy2']].sum().apply(np.exp)  # $ 1 return
```

```python
data[['returnsbh','strategy','strategy2']].std()*np.sqrt(252)
```

strategy 2 works better than both strategy 1 and returns from buy and hold, while decreasing std
